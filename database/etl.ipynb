```python
import mysql.connector

db_data={
    "host":"localhost",
    "user":"root",
    "password":"xxxxx",
    "database":"ecom_db"
}
conn = mysql.connector.connect(**db_data)
try:
    if conn.is_connected():
        print("Database Connected")
        cursor = conn.cursor()

        # update sales fact table
        sql_query = """INSERT INTO sales_fact(
        order_id,
        product_id,
        customer_id,
        warehouse_id,
        qty,
        unit_price, 
        total_amount,
        order_date
        )
        SELECT 
            o.order_id,
            oi.product_id,
            o.customer_id,
            oi.warehouse_id,
            oi.qty,
            p.price_per_qty,
            oi.tot_price,
            o.order_date
        FROM
            orders o
        JOIN order_items oi
            ON o.order_id = oi.order_id
        JOIN product p 
            ON p.product_id = oi.product_id
        WHERE DATE(o.order_date) = CURDATE()"""
        cursor.execute(sql_query)
        print("Data Loaded")
        cursor.close()
        conn.commit()
        conn.close()
except mysql.connector.Error as e:
    f"Data Loading Failed - {e}"
```
